# Sqaure Mile Plots for Cities

The sqaure mile plots give the road network for a given city and prominent places could be identified from the networks.

```python
import osmnx as ox
from IPython.display import Image
%matplotlib inline
ox.config(log_file=True, log_console=True, use_cache=True)
```

```python
# configure the inline image display
img_folder = 'images'
extension = 'png'
size = 240
dpi = 40
```

## Ulaanbaatar

```python
place = 'Ulaanbaatar'
## (47.920901, 106.917269) coordinates for Government Palace, Mongolia
point = (47.920901, 106.917269)
fig, ax = ox.plot_figure_ground(point=point, filename=place, default_width=5, network_type='drive', dpi=dpi)
Image('{}/{}.{}'.format(img_folder, place, extension), height=size, width=size)
```
